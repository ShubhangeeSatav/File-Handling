```python
#Q. Write a function in python to read the content from 
# a text file "poem.txt" line by line and display the same on screen.

import os
def read_file():
    if os.path.exists("poem.txt"):
        with open("poem.txt", "r") as f:
            for x in f:
                print(x)
            f.close()
    else:
        print("The file does not exist")

read_file()
```

```python
#Q. Write a function in python to count the number of lines 
# from a text file "story.txt" which is not starting with an alphabet "T". 

# Example: If the file "story.txt" contains the following lines: A boy is playing there.
# There is a playground.
# An aeroplane is in the sky.
# The sky is pink.
# Alphabets and numbers are allowed in the password.
# The function should display the output as 3

import os
def read_story():
    if os.path.exists("story.txt"):
        with open("story.txt", "r") as f:
            count = 0
            for x in f:
                if x[0] not in 'T':
                    count += 1
            f.close()
            print("Number of lines not starting with 'T'=", count)        
    else:
        print("The file does not exist")
        
read_story()
```

```python
#Q. Write a function in Python to read lines from a text file "notes.txt". 
# Your function should find and display the occurrence of the word "the".

# For example: If the content of the file is:
# "India is the fastest-growing economy. India is looking for more investments around the globe. The whole world is looking at India as a great market. Most of the Indians can foresee the heights that India is capable of reaching."

# The output should be 5.

import os
def notes_read():
    if os.path.exists("notes.txt"):
        with open("notes.txt", "r") as f:
            data = f.read()
            word = data.split()
            count = 0
            for x in word:
                if x == 'the'  or x == 'The':
                    count += 1
            f.close()
            print("The occurrence of the word 'the' is : ", count)        
    else:
        print("The file does not exist")
            
notes_read()              
```

```python
#Q. Write a function display_words() in python to read lines from a text file "story.txt", 
# and display those words, which are less than 4 characters.

import os
def display_words():
    if os.path.exists("story.txt"):
        with open("story.txt", "r") as f:
            data = f.read()
            words = data.split()
            count = 0
            for x in words:
                if len(x) < 4:
                    count += 1
            f.close()
            print("The Words which are less than 4 characters :", count)
    else:
        print("The File does not exist")

display_words()
```

```python
#Q. Write a function in Python to count uppercase character in a text file.

import os
def count_article():
    if os.path.exists("article.txt"):
        with open("article.txt", "r") as f:
            data = f.read()
            words = data.split()
            count = 0
            for x in words:
                if x.isupper():
                    count +=1
            f.close()
            print("count uppercase character in a text file is: ", count)
    else:
        print("The File does not exist")
        
count_article()
```

```python
#Q. A text file named "matter.txt" contains some text, which needs to be displayed such that 
# every next character is separated by a symbol "#". 
# Write a function definition for hash_display() in Python that would display 
# the entire content of the file matter.txt in the desired format.

# Example :
# If the file matter.txt has the following content stored in it :
# THE WORLD IS ROUND

# The function hash_display() should display the following content :
# T#H#E# #W#O#R#L#D# #I#S# #R#O#U#N#D#

import os
def hash_display():
    if os.path.exists("matter.txt"):
        with open("matter.txt", "r") as f:
            data = f.read()
            for x in data:
                print(x, end='#')
            f.close()
    else:
        print("The File does not exist")
        
hash_display()          
```

```python
#Q. Aditi has used a text editing software to type some text. After saving the article as WORDS.TXT, 
# she realised that she has wrongly typed alphabet J in place of alphabet I everywhere in the article.
# Write a function definition for JTOI() in Python that would display the corrected version of entire content 
# of the file WORDS.TXT with all the alphabets "J" to be displayed as an alphabet "I" on screen.

# Note: Assuming that WORD.TXT does not contain any J alphabet otherwise.

# Example:
# If Aditi has stored the following content in the file WORDS.TXT:
# WELL, THJS JS A WORD BY JTSELF. YOU COULD STRETCH THJS TO BE A SENTENCE
# The function JTOI() should display the following content:
# WELL, THIS IS A WORD BY ITSELF. YOU COULD STRETCH THIS TO BE A SENTENCE

   
import os
def JTOI():
    if os.path.exists("WORD.txt"):
        with open("WORD.txt", "r") as f:
            data = f.read()
            for x in data:
                if x == 'J':
                    print("I", end="")
                else:
                    print(x, end="")
            f.close()
    else:
        print("The File does not exist")
        
JTOI()
```

```python
#Q. A binary file "Book.dat" has structure [BookNo, Book_Name, Author, Price].
# i. Write a user defined function createFile() to input data for a record and add to Book.dat.
# ii. Write a function countRec(Author) in Python which accepts the Author name as parameter 
# and count and return number of books by the given Author are stored in the binary file "Book.dat"

import pickle
def createFile():
    file = open("book.dat", "ab")
    BookNo = int(input("Enter a Book Number :"))
    Book_Name = input("Enter a Book Name :")
    Author = input("Enter Author Name :")
    Price = int(input("Enter Price :"))
    record = [BookNo, Book_Name, Author, Price]
    pickle.dump(record, file)
    file.close()
    
def countRec(Author):
    file = open("book.dat", "rb")
    count = 0
    try:
        while True:
            record = pickle.load(file)
            if record[2] == Author:
                count +=1
    except EOFError:
        pass
    return count
    file.close()
    
def my_function():
    while True:
        createFile()
        choice = input("\nAdd more records (y/n)? ")
        if choice in 'Nn':
            break
    Author = input("\nEnter Author name to search :")
    n = countRec(Author)
    print("Number of books are :", n)
    
my_function()
```

```python
#Q. A binary file "STUDENT.DAT" has structure (admission_number, Name, Percentage). 
# Write a function count_rec() in Python that would read contents of the file "STUDENT.DAT" 
# and display the details of those students whose percentage is above 75. 
# Also display number of students scoring above 75%

import pickle
def createFile():
    file = open("student.dat", "ab")
    admission_number = int(input("Enter Admission Number of Student: "))
    Name = str(input("Enter Name of Student: "))
    Percentage = float(input("Enter Percentage of Student: "))
    record = [admission_number, Name, Percentage]
    pickle.dump(record, file)
    file.close()

def count_rec():
    file = open("student.dat", "rb")
    count = 0
    try:
        print("\nStudent Details of more than 75% are : ")
        while True:
            record = pickle.load(file)
            if record[2] > 75:
                print("\nStudent Admission Number is :", record[0])
                print("Student Name is :", record[1])
                print("Percentage is :", record[2])
                count +=1
    except EOFError:
        pass
    print('\nNumber of students having more than 75% are: ', count)
    file.close()

def my_function():
    while True:
        createFile()
        choice = input("\nAdd more records (y/n)? ")
        if choice in 'Nn':
            break
    count_rec()
    
my_function()    
```

```python
#Q. A binary file school.dat has structure(rollno, name, class, fees)

# Write a definition for function total_fees( ) that reads each object of file and calculate 
# the total fees of students and display the same.

import pickle
def createSchoolFile():
    file = open("school.dat", "ab")
    Rollno = int(input("Enter Roll Number of Student: "))
    Name = str(input("Enter Name of Student: "))
    Class = input("Enter Class of Student: ")
    Fees = input("Enter Fees of Student: ")
    record = [Rollno, Name, Class, Fees]
    pickle.dump(record, file)
    file.close()

def total_fees():
    file = open("school.dat", "rb")
    total = 0
    try:
        while True:
            record = pickle.load(file)
            total = total + int(record[3])
    except EOFError:
        pass
    print('\nTotal Fees: ', total)
    file.close()

def my_function():
    while True:
        createSchoolFile()
        choice = input("\nAdd more records (y/n)? ")
        if choice in 'Nn':
            break
    total_fees()
          
my_function()
```

```python
#Q. A binary file players.dat, containing records of following list format: [code, name, country and total runs]

# 1. Write a python function that display all records where player name starts from 'A'
# 2. Write a python function that accept country as an argument and count and display 
#     the number of players of that country.
# 3. Write a python function that add one record at the end of file.


import pickle
def createPlayersFile():
    file = open("players.dat", "ab")
    Code = int(input("Enter Code of Player: "))
    Name = str(input("Enter Name of Player: "))
    Country = input("Enter Country of Player: ")
    Total_runs = input("Enter Total Runs of Player: ")
    record = [Code, Name, Country, Total_runs]
    pickle.dump(record, file)
    file.close()

def PlayersNameStartsFromA():
    file = open("players.dat", "rb")
    print("\nPlayer name starts from 'A' Details: ")
    try:
        while True:
            record = pickle.load(file)
            #print(record)
            if record[1].startswith('A'):
                print("\nPlayer Code is :", record[0])
                print("Player Name is :", record[1])
                print("Player Country is :", record[2])
                print("Player Total Runs is :", record[3])
    except EOFError:
        pass
    file.close()
    
def countCountryPlayers(country):
    file = open("players.dat", "rb")
    count = 0
    try:
        while True:
            record = pickle.load(file)
            if record[2] == country:
                count += 1
    except EOFError:
        pass
    print('\nThe number of players of that country: ', count)
    file.close()
    
def my_function():
    while True:
        createPlayersFile()
        choice = input("\nAdd more records (y/n)? ")
        if choice in 'Nn':
            break
    country = input("\nEnter country to search :")
    countCountryPlayers(country)
    PlayersNameStartsFromA()
          
my_function()
```

```python
#Q. Given a binary file game.dat, containing records of following list format: [game_name, participants]

# Write a function in Python that would read contents from the file game.dat and 
# creates a file named basket.dat copying only those records from game.dat where the game name is "Basket Ball"

import pickle
def createGameFile():
    file = open("game.dat", "ab")
    game_name = input("Enter Game Name: ")
    participants = input("Enter Number of Participants: ")
    record = [game_name, participants]
    pickle.dump(record, file)
    file.close()

def basket_record():
    file1= open("game.dat", "rb")
    file2= open("basket.dat", "wb")
    try:
        while True:
            record = pickle.load(file1)
            if record[0] == 'Basket Ball':
                pickle.dump(record, file2)
                print("\nNumber of Basket Ball participants : ", record[1])
    except EOFError:
        pass
    file1.close()
    file2.close()

def my_function():
    while True:
        createGameFile()
        choice = input("\nAdd more records (y/n)? ")
        if choice in 'Nn':
            break
    basket_record()
          
my_function()
```
